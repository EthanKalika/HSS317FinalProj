First lets load the data and import libraries.

```python
import pandas as pd

data1 = pd.read_csv("data/nba_2022-23_all_stats_with_salary.csv")

print(data1.shape)
data1.head()
```

Now lets drop the columns that we will not use.

```python
data1 = data1.drop(columns = ['Unnamed: 0', 'Player Name', 'Team', 'GP', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '2P', '2PA', 'eFG%', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS', '3PAr', 'FTr', 'USG%', 'OWS', 'WS', 'DWS', 'OBPM', 'BPM', 'VORP', 'DBPM'])
print(data1.shape)
data1.head()
```

Now lets deal with the position categorical variable

```python
data1['SG'] = 0
data1['PF'] = 0
data1['C'] = 0
data1['SF'] = 0
data1['PG'] = 0

for index, row in data1.iterrows():
  currPos = row['Position'].split("-")
  for i in currPos:
    if i == 'SG':
      data1.at[index, 'SG'] = 1
    elif i == 'PF':
      data1.at[index, 'PF'] = 1
    elif i == 'C':
      data1.at[index, 'C'] = 1
    elif i == 'SF':
      data1.at[index, 'SF'] = 1
    elif i == 'PG':
      data1.at[index, 'PG'] = 1

data1 = data1.drop(columns = ['Position'])

data1.head()
```

Now lets normalize the data.

```python
maxSal = data1['Salary'].max()
data1Norm = data1/data1.max() * 10

for index, row in data1Norm.iterrows():
  data1Norm.at[index, 'Salary'] = row['Salary'] * maxSal / 1000000

print(data1Norm.head())
```

Lets add an age^2 column. This will help account for the fact that more extreme ages result in less favorable pay.

```python
data1Norm['Age^2'] = data1Norm['Age'] ** 2
data1Norm = data1Norm.fillna(0)
data1Norm.to_csv("data/finaldata.csv")
data1Norm.head()
```
